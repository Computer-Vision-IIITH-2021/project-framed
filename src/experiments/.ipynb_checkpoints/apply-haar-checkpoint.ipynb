```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import torch
import torch.nn as nn
```

```python
from harr import *
```

```python
LL, LH, HL, HH = get_harr_wav(3)
```

```python
def get_convolved_image(im, filt):
    
    im_con = filt(torch.tensor(im[None, ...]).permute(0, 3, 1, 2).type('torch.FloatTensor'))
    im_ret = np.transpose(im_con.detach().numpy(), (0, 2, 3, 1))[0, :, :, 0]
    im_ret = im_ret/np.max(im_ret)
    
    return im_ret
```

```python
def plot_convolutions(im1, im2, im3, im4):
    
    fig, ax  = plt.subplots(2,2, figsize = (10,10))

    ax[0][0].imshow(im1, cmap='gray')
    ax[0][0].set_title("LL Haar filter")

    ax[0][1].imshow(im2, cmap='gray')
    ax[0][1].set_title("LH Haar filter")

    ax[1][0].imshow(im3, cmap='gray')
    ax[1][0].set_title("HL Haar filter")

    ax[1][1].imshow(im4, cmap='gray')
    ax[1][1].set_title("HH Haar filter")
    
```

```python
louvre = cv2.imread("../images/content/louvre.png")
louvre = cv2.cvtColor(louvre, cv2.COLOR_BGR2RGB)
plt.imshow(louvre)
```

```python
LL_louvre = get_convolved_image(louvre, LL)
LH_louvre = get_convolved_image(louvre, LH)
HL_louvre = get_convolved_image(louvre, HL)
HH_louvre = get_convolved_image(louvre, HH)
```

```python
plot_convolutions(LL_louvre, LH_louvre, HL_louvre, HH_louvre)
```

```python
cliff = cv2.imread("../images/content/cliff.png")
cliff = cv2.cvtColor(cliff, cv2.COLOR_BGR2RGB)
plt.imshow(cliff)
```

```python
LL_cliff = get_convolved_image(cliff, LL)
LH_cliff = get_convolved_image(cliff, LH)
HL_cliff = get_convolved_image(cliff, HL)
HH_cliff = get_convolved_image(cliff, HH)
```

```python
plot_convolutions(LL_cliff, LH_cliff, HL_cliff, HH_cliff)
```
